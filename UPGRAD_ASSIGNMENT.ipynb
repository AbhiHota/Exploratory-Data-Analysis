--**Defining function to treat outliers of a column**

````python
def outlier_treatment(column_name):
    q1=inp1_r[[column_name]].describe().loc["25%"][column_name]
    q3=inp1_r[[column_name]].describe().loc["75%"][column_name]
    iqr=q3-q1
    up_bd=q3+1.5*iqr
    lo_bd=q1-1.5*iqr
    inp1_r[column_name]=np.where(inp1_r[column_name]>up_bd,up_bd,inp1_r[column_name])
    inp1_r[column_name]=np.where(inp1_r[column_name]<lo_bd,lo_bd,inp1_r[column_name])
    return
````

-- **Defining function to impute null values of a column**

````python
def null_treatment(column_name):
    inp1_r[column_name]=inp1_r[column_name].fillna(inp1_r[column_name].median())
    return
````
